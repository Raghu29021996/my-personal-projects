```python
import numpy as np
import scipy
from scipy import signal
from skimage import io, color
from skimage import exposure
import matplotlib.pyplot as plt
%matplotlib inline


img = io.imread('apple1.jpg')  # Load the image
print('\n First 5 columns and rows of the image_sharpen matrix: \n', img[:3,0:3,0:3])
plt.imshow(img)
plt.axis('off')
plt.show()

img = color.rgb2gray(img)  # Convert the image to grayscale (1 channel)
print('\n First 5 columns and rows of the image matrix: \n', np.floor(img[150:155,250:255]*255))
plt.imshow(img,cmap=plt.cm.gray)
plt.axis('off')
plt.show()
kernel = np.array([[-1,-1,-1], 
                   [-1,8,-1],
                   [-1,-1,-1]])

image_conv = scipy.signal.convolve2d(img, kernel, 'same')
print('\n First 5 columns and rows of the image_sharpen matrix: \n', np.floor(image_conv[150:155,250:255]*255))

image_conv_equalized = exposure.equalize_adapthist(image_conv/np.max(np.abs(image_conv)), clip_limit=0.03)
plt.imshow(image_conv_equalized,cmap=plt.cm.gray)
plt.axis('off')
plt.show()
```
